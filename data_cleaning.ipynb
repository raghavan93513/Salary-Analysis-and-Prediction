# **Things to do**

1) Drop Unneccessary columns
2) Make Colums in lower case
3) Remove all text after delimiter in "job title", "location", "headquarters", "revenue" and split location into state and city attributes
4) rename columns - "type of ownership" to "ownership" and "salary estimate" to "salary($)"
5) Replace unknown with NaN in "ownership"
6) Replace -1 with False in "easy apply"
7) Replace TRUE with True in "easy apply"
8) Replace -1 with NaN in "sector"
9) Replace -1 with NaN in "size"
10) Replace -1 with NaN in "founded"
11) Replace -1 with NaN in "rating"
12) Replace -1 with NaN in "company name"
13) Drop the rows containing NA and reset the indexing
14) Remove the dollor sign, 'k' and the text in "salary($)"
15) Make 3 new columns "min salary", "max salary" and "avg salary" using "salary($)"
16) Typecase the values in "founded" to int and find the "age" of the company
17) Format the "company name" attribute by removing unwanted text
18) Make an attribute to show "headquarters" is same as the city
19) Check if the job involves python, aws or excel. Even find the length of description. Then drop "job description"
20) Change "United Kingdom" in State to "UK".
21) Make a "seniority" attribute to show if the position is senior or junior - sr = senior = lead = principal and jr = junior
22) Make a "job simplification" attribute - says if the attribute involves data scientist, data engineer, machine learning, manager, director. Else NA
23) Make an attribute for number of competitors and drop the "competitors" attribute

## 1

```python
df.drop(['Unnamed: 0','index','Industry'],axis=1,inplace=True)
df
```

```python
df.columns
```

## 2

```python
df.columns=map(str.lower,df.columns)
df.columns
```

```python
df['job title'].value_counts()
```

```python
df['job title']
```

## 3

```python
df['job title']=df['job title'].apply(lambda x:x.join(x.split(',')[:1]))
df['job title']=df['job title'].apply(lambda x:x.join(x.split('/')[:1]))
df['job title']=df['job title'].apply(lambda x:x.join(x.split('-')[:1]))
df['city']=df['location'].apply(lambda x:x.split(',')[0])
df['state']=df['location'].apply(lambda x:x.split(',')[1])
df.drop(['location'],axis=1,inplace=True)
df['headquarters']=df['headquarters'].apply(lambda x:x.join(x.split(',')[:1]))
df['revenue']=df['revenue'].apply(lambda x: x.split('/')[0] if x[0]!="$" else x)
df
```

## 4

```python
df.rename(columns={'type of ownership':'ownership','salary estimate':'salary($)'},inplace=True)
df.columns
```

## 5

```python
df['ownership']=df['ownership'].replace(['Unknown'],np.nan)
df['ownership']
```

## 6 and 7

```python
df['easy apply']=df['easy apply'].replace(['-1'],'False')
df['easy apply']=df['easy apply'].replace(['TRUE'],'True')
df['easy apply']
```

## 8 to 12

```python
df['sector']=df['sector'].replace(["-1"],np.nan)
df['size']=df['size'].replace(["-1"],np.nan)
df['founded']=df['founded'].replace([-1],np.nan)
df['rating']=df['rating'].replace(["-1"],np.nan)
df['company name']=df['company name'].replace(["-1"],np.nan)
df
```

## 13

```python
df=df.dropna(how='any',axis=0) #removes all the rows with NA values
df=df.reset_index(drop=True) #resets all the index from the start
df
```

## 14

```python
df['salary($)']=df['salary($)'].apply(lambda x:x.split()[0])
df['salary($)']=df['salary($)'].apply(lambda x:x.replace('$',''))
df['salary($)']=df['salary($)'].apply(lambda x:x.replace('K',''))
df['salary($)']
```

## 15

```python
df['salary($)'] = df['salary($)'].apply(lambda x : x.split("(")[0])
df['min salary'] = df['salary($)'].apply(lambda x : int(x.split("-")[0]))
df['max salary'] = df['salary($)'].apply(lambda x : int(x.split("-")[1]))
df['avg salary'] = (df['min salary']+df['max salary'])/2 
df.drop(['salary($)'],axis=1,inplace=True)
df
```

## 16


```python
df['founded']=df['founded'].astype(int)
df['age'] = df['founded'].apply(lambda x: 2022-x)
df
```

## 17

```python
df['company name'] = df.apply(lambda x : x['company name'] if x['rating'] < 0 else x['company name'][:-4], axis=1)
df['company name']
```

## 18

```python
df['headquarters same city'] = df.apply(lambda x: "yes" if x['headquarters'] == x['city'] else "no", axis=1)
df
```

```python
df['headquarters same city'].value_counts()
```

## 19

```python
df['python involved'] = df['job description'].apply(lambda x : "yes" if "python" in x.lower() else "no")
df['aws involved'] = df['job description'].apply(lambda x : "yes" if "aws" in x.lower() else "no")
df['excel involved'] = df['job description'].apply(lambda x : "yes" if "excel" in x.lower() else "no")
df['tableau involved'] = df['job description'].apply(lambda x : "yes" if "tableau" in x.lower() 
                                                                      or "business intelligence" in x.lower() 
                                                                      else "no")
df['powerBI involved'] = df['job description'].apply(lambda x : "yes" if "powerbi" in x.lower() 
                                                                      or "power bi" in x.lower() 
                                                                      or "business intelligence" in x.lower() 
                                                                      or "mspower" in x.lower()
                                                                      else "no")

df['description length'] = df['job description'].apply(lambda x : len(x))
df
```

```python
print(df['python involved'].value_counts())
print("*************************************")
print(df['aws involved'].value_counts())
print("*************************************")
print(df['excel involved'].value_counts())
print("*************************************")
print(df['tableau involved'].value_counts())
print("*************************************")
print(df['powerBI involved'].value_counts())
```

## 20

```python
df['state'].value_counts()
```

```python
df['state'] = df['state'].apply(lambda x: "UK" if "United Kingdom" in x else x.strip()) #strip removes any white spaces
df['state'].value_counts()
```

## 21

```python
df['seniority'] = df['job title'].apply(lambda x : "senior" if "sr" in x.lower() or "senior" in x.lower() or "lead" in x.lower() or "principal" in x.lower() 
                                              else "junior" if "jr" in x.lower() or "junior" in x.lower() 
                                              else "none")
df['seniority'].value_counts()
```

## 22

```python
df['job simplification'] = df['job title'].apply(lambda x : "data scientist" if "data scientist" in x.lower() 
                                                        else "data engineer" if "data engineer" in x.lower()
                                                        else "machine learning" if "machine learning" in x.lower()
                                                        else "manager" if "manager" in x.lower()
                                                        else "director" if "director" in x.lower()
                                                        else "none")
df['job simplification'].value_counts()
```

## 23

```python
df['no of competitors'] = df['competitors'].apply(lambda x : len(x.split(",")) if x != '-1' else "No data")
df.drop(['competitors'],axis=1,inplace=True)
```
